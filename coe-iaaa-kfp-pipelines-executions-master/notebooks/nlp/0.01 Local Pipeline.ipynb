```python
from aikfp.ingestamg_simulation import *
from aikfp.src.automl.nlp_utils import clean_text
pd.set_option('display.max_columns',None)
```

```python
df=pd.concat([pd.read_csv(x) for x in ['data_entries/nlp_opendata_backup_v1.csv','data_entries/nlp_opendata_backup_v2.csv','data_entries/nlp_opendata_backup_v3.csv']])
# df=df.sample(10000)
df['id_review']=[f'R_{x}' for x in range(len(df))]
df['score']=df['score'].map({1:0,2:1,3:1,3:1,4:1,5:2})
df['review']=df['review'].apply(clean_text)
df.drop_duplicates(subset=['review'],inplace=True)
def revise(x):
    try:
        x.lower()
        return 1
    except: return 0
df['bug']=df['review'].apply(revise)
df=df[df['bug']!=0]
```

```python
ml=automl_training('nlp_sentiment',df)
```
